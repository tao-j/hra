```python
from algo_func import *
from data_helper import *
%matplotlib inline

data_seeds = [1313, 3838, 6262, 1338, 1362, 3862, 6238, 6213, 3813, 13, 38, 62]
betas = range(16)
acc_list = []
result_pack = []
for idx, ds in enumerate(data_seeds):
    this_seed_acc_list = []
    this_seed_rst_list = []
    for i in range(1):
        kwarg = {
            'data_seed': ds,
            'shrink_b':  i, # (16-i) * 6 +
            'n_items': 10,
            'n_judges': 2,
            'n_pairs': 1000,
            'beta_gen_func': 'power'
        }
        print(kwarg)
        data_pack = generate_data(**kwarg)

        rank, acc = train_func_torchy(data_pack, init_seed=ds, debug=True, verbose=0,
                                      result_pack=this_seed_rst_list,
                                      algo='simple', max_iter=200, lr=5*1e-3, opt_func=torch.optim.SGD)
        print(rank, acc)
        this_seed_acc_list.append(acc)
        
        print('------------------')
    acc_list.append(this_seed_acc_list)
    result_pack.append(this_seed_rst_list)
# seed 1:2 not working
```

```python
clip = 50
pd_dict = {}
key = ["betas", "ranking error rate", "\hat{s} deviation from s", "%%/100 deviation from beta (clipped at %.2f)" % clip]
val = betas, [1 - i[0] for i in acc_list[0]], [rs[0] for rs in result_pack[0]], [rs[1] if rs[1] < clip else clip for rs in result_pack[0]]
for k, v in zip(key, val):
    pd_dict[k] = v
    
eval_df = pd.DataFrame.from_dict(pd_dict)
eval_df.set_index("betas")
ax = eval_df.plot(x="betas",logx=True)
ax.xaxis.grid()
ax.yaxis.grid()
```

```python
clip = 0.3
clip_err = 0.3
pd_dict = {}
for i_b in range(len(data_seeds)):
    key = ["betas", "ranking error rate", "\hat{s} deviation from s", "%%/100 deviation from beta (clipped at %.2f)" % clip]
    val = betas, \
    [1 - i[0] if (1 - i[0] < clip_err) else clip_err for i in acc_list[i_b]], [rs[0] for rs in result_pack[i_b]], [rs[1] if rs[1] < clip else clip for rs in result_pack[i_b]]
    for k, v in zip(key, val):
        pd_dict[k] = v

    eval_df = pd.DataFrame.from_dict(pd_dict)
    eval_df.set_index("betas")
    ax = eval_df.plot(x="betas",logx=False)
    ax.xaxis.grid()
    ax.yaxis.grid()
```
